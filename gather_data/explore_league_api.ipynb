```python
def get_all_ranked_league_match_by_uuid(puuid,start_date,region,index=0,match_ids=[]):
    url = f'https://{region}.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?startTime={start_date}&queue=420&type=ranked&start={index}&count=100'
    headers = {'X-Riot-Token': API_KEY}
    response = requests.get(url, headers=headers)
    res=response.json()
    match_ids.extend(res)
    if len(res) == 100 :
        index +=100
        get_all_ranked_league_match_by_uuid(puuid,start_date,region,index,match_ids)
    return match_ids



match_ids =get_all_ranked_league_match_by_uuid(puuid=puuid,start_date=timestamp_season14_start,region='asia')
len(match_ids)
```

```python
def get_match_info_by_match_id(match_id,region):
    url= f'https://{region}.api.riotgames.com/lol/match/v5/matches/{match_id}'
    headers = {'X-Riot-Token': API_KEY}
    response = requests.get(url, headers=headers)
    return response.json()

match_data= get_match_info_by_match_id(match_id=match_ids[0],region='asia')
```

```python
def parse_match_data(match_data):
    '''
    match_data  : Json match data coming from match/v5/matches/by match_id from riot official api

    '''
    list_participant = match_data["metadata"]["participants"]
    list_info = match_data["info"]["participants"]
    dic_match_data={}
    for info_player in list_info :
        participant = info_player["puuid"]

        dic_match_data[participant] = {}
        #print(dic_match_data.keys())

        dic_match_data[participant]["riotIdGameName"] = info_player["riotIdGameName"] # nu
        dic_match_data[participant]["riotIdTagline"] = info_player["riotIdTagline"] #nu
        dic_match_data[participant]["championName"] = info_player["championName"] 
        dic_match_data[participant]["championId"] = info_player["championId"] #nu maybe
        dic_match_data[participant]["teamPosition"] = info_player["teamPosition"]
        dic_match_data[participant]["role"] = info_player["role"] #nu
        dic_match_data[participant]["individualPosition"] = info_player["individualPosition"] #nu
        dic_match_data[participant]["lane"] = info_player["lane"] #nu
        dic_match_data[participant]["summonerLevel"] = info_player["summonerLevel"]
        dic_match_data[participant]["kills"] = info_player["kills"] #nu
        dic_match_data[participant]["deaths"] = info_player["deaths"] #nu
        dic_match_data[participant]["assists"] = info_player["assists"] #nu
        dic_match_data[participant]["teamId"] = info_player["teamId"] #nu maybe
        dic_match_data[participant]["win"] = info_player["win"] 

    return dic_match_data

parsed_match = parse_match_data(match_data)
```
